```python
numeric_columns = ['avg_vote', 'metascore', 'duration']
```

```python
cleaned_df = df.dropna(subset=['genre', 'director', 'actors', 'budget', 'usa_gross_income'])
print(cleaned_df)
```

```python
cleaned_df = df[numeric_columns].dropna()
```

```python
if 'genre' in df.columns:
    # Perform cleaning and analysis steps here
    cleaned_df = df[['genre', 'avg_vote']]  # Example: Retaining only 'genre' and 'avg_vote' columns
    # Further analysis or visualization
else:
    print("'genre' column not found in dataset. Check your data loading or cleaning steps.")
```

```python
cleaned_df.describe()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(cleaned_df['avg_vote'], bins=20, kde=True)  # Example: Histogram of average votes
plt.title('Distribution of Average Votes')
plt.xlabel('Average Vote')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(y='genre', data=cleaned_df, order=cleaned_df['genre'].value_counts().index[:10])
plt.title('Top 10 Movie Genres')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

```python
correlation_matrix = cleaned_df.corr()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of IMDb Movie Features')
plt.show()
```

```python
avg_ratings_by_genre = cleaned_df.groupby('genre')['avg_vote'].mean().sort_values(ascending=False).head(10)
print(avg_ratings_by_genre)
```
